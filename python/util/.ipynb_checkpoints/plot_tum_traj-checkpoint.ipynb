```python
import plotly.graph_objs as go
from plotly.subplots import make_subplots
import plotly.offline as py
import numpy as np
```

```python
# Dateiname der Trajektoriedaten
camera_data = '/mnt/hdd02/ba_copter_videos/Flight_10/dso_traj.txt'
```

```python
# Listen für Delta-Werte
cam_delta_twb_x = []
cam_delta_twb_y = []
cam_delta_twb_z = []

# Trajektoriedaten einlesen
with open(camera_data, 'r') as f:
    for line in f:
        teile = line.split()
        cam_delta_twb_x.append(float(teile[1]))
        cam_delta_twb_y.append(float(teile[2]))
        cam_delta_twb_z.append(float(teile[3]))

# Erstelle interaktiven 3D-Plot
fig = make_subplots(rows=1, cols=1, specs=[[{'type': 'scatter3d'}]])

# Plot hinzufügen
fig.add_trace(go.Scatter3d(x=cam_delta_twb_x, y=cam_delta_twb_y, z=cam_delta_twb_z, mode='lines', name='Trajectory'))

# Layout anpassen
fig.update_layout(title='Trajectory',
                  scene=dict(xaxis=dict(title='Z-Coordinate'),  # Ändere die Beschriftung der Achsen
                             yaxis=dict(title='Y-Coordinate'),
                             zaxis=dict(title='X-Coordinate')),  # Ändere die Beschriftung der Achsen
                  margin=dict(l=0, r=0, t=30, b=0))

# Plot anzeigen
# py.plot(fig, filename='/home/bastian/uni/visual_odometry_playground/orb_slam3/orb_slam3_ws/getrackter_pfad.html')
py.iplot(fig)
```
